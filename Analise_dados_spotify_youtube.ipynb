# Analise de dados do Spotify e do YouTube

Dataset: https://www.kaggle.com/datasets/salvatorerastelli/spotify-and-youtube

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Spotify_Youtube.csv')
df.head()
```

Logo a partir dessas visualizações, conseguimos notar que o dataset possui 20718 linhas e 28 colunas.
Ele também possui algumas váriaveis nulas

Fazendo algumas análises básicas

```python
# Top 20 artistas mais ouvidos no Spotify
top_20_artists_spotify = df.groupby(['Artist'])['Stream'].sum().sort_values(ascending=False).head(20)
top_20_artists_spotify
```

```python
plt.figure(figsize=(10, 6))
top_20_artists_spotify.sort_values().plot(kind='barh')
plt.title('Top 20 artistas mais ouvidos no Spotify')
plt.xlabel('Quantidade de streams')
plt.ylabel('Artistas')
plt.show()
```

Análisando correlação entre os dados, onde pelo coeficiente de Pearson, temos que:

ρ = 0,9 a 1 (positivo ou negativo): correlação muito forte;

ρ = 0,7 a 09 (positivo ou negativo): correlação forte;

ρ = 0,5 a 0,7 (positivo ou negativo): correlação moderada;

ρ = 0,3 a 0,5 (positivo ou negativo): correlação fraca;

ρ = 0 a 0,3 (positivo ou negativo): não possui correlação.

```python
correlation = df.corr(numeric_only=True)
correlation
```

```python
plt.figure(figsize = (28,10))
sns.heatmap(correlation,
            xticklabels=correlation.columns,
            yticklabels=correlation.columns,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.4    )
```

Logo podemos ver que não existe uma correlação tão forte de streams relacionadas com atributos musicais

```python
musical_attributes = ['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness',
       'Acousticness', 'Instrumentalness', 'Liveness', 'Valence']

plt.figure(figsize=(15, 10))
for i, variable in enumerate(musical_attributes, 1):
    plt.subplot(2, len(musical_attributes)//2 + len(musical_attributes) % 2, i)
    sns.histplot(data=df, x=variable, kde=True)
    plt.title(f'Histograma de {variable}')

plt.tight_layout()
plt.show()
```

Fazendo uma análise da relação de streams do Spotify com os atributos musicais

```python
plt.figure(figsize=(15, 10))
for i, variable in enumerate(musical_attributes, 1):
    plt.subplot(2, len(musical_attributes)//2 + len(musical_attributes) % 2, i)
    sns.scatterplot(data = df, x = variable, y = df['Stream'])
    plt.title(f'Scatterplot de {variable}')

plt.tight_layout()
plt.show()
```

Com essa análise conseguimos perceber: que as pessoas preferem músicas que não estão em áudio acustico, que não sejam apenas instrumentais, que não sejam músicas ao vivo e preferem mais músicas cantadas.

Fazendo a mesma análise mas com as views do Youtube

```python
plt.figure(figsize=(15, 10))
for i, variable in enumerate(musical_attributes, 1):
    plt.subplot(2, len(musical_attributes)//2 + len(musical_attributes) % 2, i)
    sns.scatterplot(data = df, x = variable, y = 'Views')
    plt.title(f'Scatterplot de {variable}')

plt.tight_layout()
plt.show()
```

```python
differences_youtube_spotify = df.groupby('Artist')[['Stream', 'Views']].sum()
differences_youtube_spotify
```
